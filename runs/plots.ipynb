```python
import matplotlib.pyplot as plt
```

```python
throughputs = {"n=5,c=5": 38.14612848132771,
               "n=5,c=10": 41.11840625079959,
               "n=5,c=20": 43.91754194350042,
               "n=10,c=5": 41.25551275439670,
               "n=10,c=10": 46.15554945646096,
               "n=10,c=20": 50.25185930384198}
keys = list(throughputs.keys())
values = list(throughputs.values())

plt.figure(figsize=(10, 6))
plt.bar(keys, values, color='royalblue',width=0.5) 
plt.xlabel('n: Nodes, c: Capacity')
plt.ylabel('Throughput')
plt.xticks(rotation=45, ha='right') 
plt.tight_layout()
plt.ylim(35, 51) 
plt.show()
```
